# `Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.`

`Seaborn is a popular Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics. Here are five types of plots you can create using Seaborn, along with their typical use cases:`



- 1. scatterplot:- Scatter plots are used to visualize the relationship between two numerical variables. Each data point is represented as a point on the plot, with one variable along the x-axis and the other along the y-axis.


- 2. Histograms display the distribution of a single numerical variable by dividing it into intervals (bins) and showing the frequency of data points in each bin.


- 3. Box plots provide a summary of the distribution of a numerical variable, showing the median, quartiles, and potential outliers.


- 4. Bar plots are used to compare categorical data by displaying the frequency or values of different categories as bars.

- 5. : Heatmaps are used to visualize the correlation between two-dimensional data by representing values with colors. They are particularly useful for exploring relationships within large datasets.

# `Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.`

```python
import seaborn as sns
fmri = sns.load_dataset('fmri')
```

```python
fmri.head()
```

```python
sns.lineplot(x=fmri.timepoint,y=fmri.signal,color='r',hue=1)
```

# `Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.`

```python
diamonds = sns.load_dataset('diamonds')
```

```python
diamonds.head()
```

```python
sns.histplot(data=diamonds,x='price',hue='cut',kde=True,color='r')
```

# `Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.`

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```
